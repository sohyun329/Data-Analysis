- HH : (같은 업종으로) 행정동의 사업체의 평균 영업기간이 "서울시"에서 "생존"한 사업체들의 평균 영업기간보다 높고, "서울시"에서 "폐업"한 사업체들의 평균영업기간보다 높은 상권 - 창업 진출입시 세심한 주의가 필요한 상권

### 시간에 따른 상권 변화 추이 분석

```python
selected_regions = ['공덕동', '아현동', '도화동', '용강동', '대흥동', '염리동', '신수동', '서강동', '서교동', '합정동', '망원제1동', '망원제2동', '연남동', '성산제1동', '성산제2동', '상암동']

# 각 지역에 대한 상권 변화 추이 분석 및 시각화
for selected_region in selected_regions:
    # 선택한 지역의 데이터 추출
    selected_region_data = df[df['행정동_코드_명'] == selected_region]

    if not selected_region_data.empty:
      # 시간에 따른 상권 변화 지표 분포 시각화
      plt.figure(figsize=(12, 6))
      sns.countplot(data=selected_region_data, x='기준_년분기_코드', hue='상권_변화_지표_명')
      plt.title(f'{selected_region} 지역의 상권 변화 추이')
      plt.xlabel('기준 년도 및 분기')
      plt.ylabel('Count')
      plt.xticks(rotation=90)
      plt.legend(title='상권 변화 지표명', loc='upper right')
      plt.tight_layout()

      # 그래프 보여주기
      plt.show()
```
